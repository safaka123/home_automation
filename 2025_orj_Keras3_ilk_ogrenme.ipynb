```python
import tensorflow as tf
from keras.utils import to_categorical
from keras.models import load_model
import pandas as pd
from sklearn.model_selection import train_test_split

dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/electrical_1.csv')
```

```python
dataset.columns = dataset.columns.str.strip()
dataset.drop(['Detector'], axis=1, inplace=True)

#Detector string olduğu için bu şekilde iptal edilir

x = dataset.drop(columns=["Status", "Time", "Reading ID"])

#Yukarıdakiler iptal edildikten sonra kalan sütunlar

print(x)
x.shape
```

```python
y = dataset["Status"]
print(y)
y.shape
```

```python
#from sklearn.model_selection import train_test_split

#Veri setinin %20'si test olarak ayrılır (test_size=0.2)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train
```

```python
y_train
```

```python
x_train
```

```python
#import tensorflow as tf
#from keras.utils import to_categorical

y_train_encoded = to_categorical(y_train)

#Veriler modelin beslenmesi için önceden kodlanır

print(y_train_encoded)
y_train_encoded.shape
```

```python
model = tf.keras.models.Sequential() #Model tanımlandı
```

```python
#Modelin katmanları tanımlanır(İlk katman - Input, İkinci katman - Gizli katman, Üçüncü - Output


model.add(tf.keras.layers.Dense(256, input_shape=(5, ), activation='sigmoid'))
model.add(tf.keras.layers.Dense(256, activation='sigmoid'))
model.add(tf.keras.layers.Dense(3, activation='softmax'))
```

```python
#Model derlenir

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#Eğitim süreci başlatılır. Epoch sayısı kadar tekrar eder

model.fit(x_train, y_train_encoded, epochs=100)
```

```python
# Ayrıca y_test'i de kodlamak gerekir

y_test_encoded = to_categorical(y_test)
print(y_test_encoded)
y_test_encoded.shape
```

```python
# Modelin ayrılmış (Önceden bilmediği) test verileri üzerinde denenip verimininin başka veri setleri kullanıldığında
# ne kadar iyi olacağını ölçülmesi. Şayet bu test seti ayrılmaz ise model verilere alışacaktır ve genele uygulanması halinde
# aynı derecede iyi bir performans sergileyemecektir

model.evaluate(x_test, y_test_encoded)
```

```python
model.save('new_model.keras') # Yeni versiyon için gerekli olursa ancak .h5 formatı hala iş görüyor
```
